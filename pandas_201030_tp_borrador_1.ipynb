```
# uso de memoria antes de la conversion de tipos (en KB)
(prospects_stage.memory_usage()/(1024)).sum()
```

```
# de ello notamos que hay muchos campos son de tipo category (además Stage que ya conociamos por la definicion del enunciado)
# de incluso muchos de ellos pueden ser reducidos a tipo boolean.
# luego de una analisis de value_counts (o eventualmente un len del mismo) para comprobar que efectivamente la supocion de que son de este tipo es correcta
# procedemos a la conversion de todos ellos
# excluimos de category fechas, precios y cualquier dato que, pese a poder ser suceptible de ser enumerado comparado con el total de lineas dada las repeticiones
# no consideramos que sea correcto convertirlos a ese tipo tanto para potencial operatoria posterior como para el anaalisis de futuros set de datos

# todo: como interpretamos a los 'None' ?

# category
prospects_stage['Stage'] = prospects_stage['Stage'].astype('category')
prospects_stage['Region'] = prospects_stage['Region'].astype('category')
prospects_stage['Territory'] = prospects_stage['Territory'].astype('category')
prospects_stage['Bureaucratic_Code'] = prospects_stage['Bureaucratic_Code'].astype('category')

prospects_stage['Billing_Country'] = prospects_stage['Bureaucratic_Code'].astype('category')
prospects_stage['Account_Type'] = prospects_stage['Account_Type'].astype('category')
prospects_stage['Opportunity_Type'] = prospects_stage['Opportunity_Type'].astype('category')
prospects_stage['Quote_Type'] = prospects_stage['Quote_Type'].astype('category')
prospects_stage['Delivery_Terms'] = prospects_stage['Delivery_Terms'].astype('category')
prospects_stage['Brand'] = prospects_stage['Brand'].astype('category')
prospects_stage['Product_Type'] = prospects_stage['Product_Type'].astype('category')
prospects_stage['Size'] = prospects_stage['Size'].astype('category')
prospects_stage['Product_Category_B'] = prospects_stage['Product_Category_B'].astype('category')
prospects_stage['Currency'] = prospects_stage['Currency'].astype('category')
prospects_stage['Last_Activity'] = prospects_stage['Last_Activity'].astype('category')
prospects_stage['Last_Modified_By'] = prospects_stage['Last_Modified_By'].astype('category')
prospects_stage['Product_Family'] = prospects_stage['Product_Family'].astype('category')
prospects_stage['Product_Name'] = prospects_stage['Product_Name'].astype('category')
prospects_stage['Prod_Category_A'] = prospects_stage['Prod_Category_A'].astype('category')
prospects_stage['Total_Amount_Currency'] = prospects_stage['Total_Amount_Currency'].astype('category')
prospects_stage['Total_Taxable_Amount_Currency'] = prospects_stage['Total_Taxable_Amount_Currency'].astype('category')
prospects_stage['TRF'] = prospects_stage['TRF'].astype('category')

prospects_stage['Delivery_Quarter'] = prospects_stage['Delivery_Quarter'].astype('category')

# boolean
prospects_stage['Pricing, Delivery_Terms_Quote_Appr'] = prospects_stage['Pricing, Delivery_Terms_Quote_Appr'].astype('bool')
prospects_stage['Pricing, Delivery_Terms_Approved'] = prospects_stage['Pricing, Delivery_Terms_Quote_Appr'].astype('bool')
prospects_stage['Bureaucratic_Code_0_Approval'] = prospects_stage['Bureaucratic_Code_0_Approval'].astype('bool')
prospects_stage['Bureaucratic_Code_0_Approved'] = prospects_stage['Bureaucratic_Code_0_Approved'].astype('bool')

# todo: este especificamenret solo tiene el valor 0... es mejor usar category?
prospects_stage['Submitted_for_Approval'] = prospects_stage['Submitted_for_Approval'].astype('bool')
		
```
