```
test_frio_frio = test_frio_frio.drop('Sales_Contract_No',1)
```

```
train_frio_frio['Stage'] = train_frio_frio['Stage'].apply(lambda x: 1 if x == 'Closed Won' else 0)
```

combierto las bariables categoricas a numeros

```
train_frio_frio_colmns_object = train_frio_frio.columns[train_frio_frio.dtypes==object].tolist() 
test_frio_frio_colmns_object = test_frio_frio.columns[test_frio_frio.dtypes==object].tolist()
```

```
new_cols_train = []
for i in train_frio_frio_colmns_object:
  new_cols_train.append(i + '1')

new_cols_test = []
for i in test_frio_frio_colmns_object:
  new_cols_test.append(i + '1')
```

```
for i in new_cols_train:
  train_frio_frio[i]=1

for i in new_cols_test:
  test_frio_frio[i]=1
```

```
for i in train_frio_frio_colmns_object:
  #train_frio_frio[i] = train_frio_frio[[i, i + '1']].groupby(i).transform(lambda x: np.sum(x) / len(train_frio_frio['ID'].tolist()))
  train_frio_frio[i] = train_frio_frio[[i, i + '1']].groupby(i).transform(lambda x: np.sum(x) / len(new_cols_train))
  train_frio_frio = train_frio_frio.drop(i + '1',1)
```

```
for i in test_frio_frio_colmns_object:
  #test_frio_frio[i] = test_frio_frio[[i, i + '1']].groupby(i).transform(lambda x: np.sum(x) / len(test_frio_frio['ID'].tolist()))
  test_frio_frio[i] = test_frio_frio[[i, i + '1']].groupby(i).transform(lambda x: np.sum(x) / len(new_cols_test))
  test_frio_frio = test_frio_frio.drop(i + '1',1)
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn import tree
import xgboost as xgb

from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = \
    train_test_split(train_frio_frio.drop(columns = 'Stage'), train_frio_frio['Stage'], test_size=0.20, random_state=42)
```

```
#rf_model = RandomForestRegressor(random_state=1)
#rf_model.fit(np.nan_to_num(X_train), y_train)
#preds = rf_model.predict(np.nan_to_num(X_test))
#rmse = np.sqrt(mean_squared_error(y_test, preds))
#print("RMSE: %f" % (rmse))
rf_model = RandomForestRegressor(random_state=1, n_estimators=100, max_depth= 250, max_features= 20)
rf_model.fit(np.nan_to_num(X_train), y_train)
preds = rf_model.predict(np.nan_to_num(X_test))
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```
from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = \
    train_test_split(train_frio_frio.drop(columns = 'Stage'), train_frio_frio['Stage'], test_size=0.2, random_state=123)
```

```
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', 
                colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```
xg_reg.fit(X_train1,y_train1)
```

```
preds1 = xg_reg.predict(X_test1)
```

```
rmse = np.sqrt(mean_squared_error(y_test1, preds1))
print("RMSE: %f" % (rmse))
```

```

xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', 
                colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 50, alpha = 25, n_estimators = 200)
xg_reg.fit(X_train1,y_train1)
preds2 = xg_reg.predict(X_test1)
rmse = np.sqrt(mean_squared_error(y_test1, preds2))
print("RMSE: %f" % (rmse))
```
