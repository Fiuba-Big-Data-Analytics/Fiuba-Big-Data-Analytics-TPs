```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder


from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV 

import matplotlib.pyplot as plt
import seaborn as sns


import xgboost as xgb
from sklearn.model_selection import train_test_split

from sklearn.model_selection import TimeSeriesSplit
```

```python
test = pd.read_csv('/home/leandro/Documentos/Organizacion de datos/Datos Santi/test_pre.csv')

train = pd.read_csv('/home/leandro/Documentos/Organizacion de datos/Datos Santi/train_pre.csv')
```

```python
train = train.drop(['Unnamed: 0', 'Opportunity_ID'], axis=1)
test = test.drop(['Unnamed: 0', 'Opportunity_ID'], axis=1)
```

```python

X_train, X_test, y_train, y_test = train_test_split(train.drop(columns = 'Stage'),train['Stage'],test_size=0.3)
```

```python
#XGBoost 

xg = xgb.XGBClassifier(learning_rate =0.05,
 n_estimators=20,
 max_depth=8,
 gamma=0)
xg.fit(X_train,y_train)
xg.score(X_test, y_test)
```

# Grid Search

# CV varios

```python
X = train.drop(columns = 'Stage')
y = train['Stage']
cv_df = pd.DataFrame(columns=['CV', 'accuracy_mean'])

for CV in range(2,51,1):

    model=xgb.XGBClassifier(learning_rate =0.3,
                     n_estimators=200,
                     max_depth=8,
                     gamma=0.5)

    tscv = TimeSeriesSplit(n_splits=CV)
    
    accuracies = cross_val_score(model, X, y, cv=tscv)

    cv_df.loc[CV-2] = [CV,accuracies.mean()]
```

```python
cv_df
```

```python
ax = cv_df.plot.line(x = 'CV', y= 'accuracy_mean', grid=True)
ax.set_xlabel("CV utilizado")
ax.set_ylabel("Score")
```
